```python
from arimasktime import *

# Define the forecast period and model name
forecast_start_date = '2023-05-01'
forecast_end_date = '2023-05-31'
model_filename = MODEL_FILENAME_DEFAULT.replace('.pkl', '')  # e.g., 'arima_model'

# Forecast river levels using the refactored forecast_river_levels function
forecast = forecast_river_levels(
    model_filename=model_filename,
    forecast_start_date=forecast_start_date,
    forecast_end_date=forecast_end_date
)

# If you wish to plot the forecast, uncomment the following lines
# Prepare data to get 'data' (if not already available)
data, _ = prepare_data()

# Get actual river levels for the forecast period if available
actual_river_levels = data['level(m)'].reindex(forecast.index)

# Call the plotting function using the updated plot_forecast function
plot_forecast(
    forecast=forecast,
    actual_river_levels=actual_river_levels
)
```
